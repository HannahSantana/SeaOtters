```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Import csv
df = pd.read_csv('database.csv')
```

```python
df.head()
```

```python
#Pull the year from the Date column
df['year'] = df['Date'].apply(lambda x: x.split("/")[-1])
```

```python
df.head()
```

```python
# Create df to count earthquakes by year
df['count'] = 1
df1 = df[['year','count']]
earthquakes_by_year = df1.groupby(['year']).sum()['count'].reset_index()

#Create df for linear regression analysis
earthquake_year = earthquakes_by_year['year'].values[:,np.newaxis]
earthquake_count = earthquakes_by_year['count'].values

earthquakes_by_year.head()
```

```python
#Plot Frequency and Linear Regression
fig = plt.figure()
plt.figure(figsize=(20, 8))

lm = linear_model.LinearRegression()

lm.fit(earthquake_year, earthquake_count)

plt.scatter(earthquake_year,earthquake_count, color='b')
plt.plot(earthquake_year, lm.predict(earthquake_year),color='r')

plt.xlabel("1965-2026", fontsize = 15)
plt.ylabel("Number of Earthquakes", fontsize = 15)
plt.title("Number of Earthquakes Since 1965", fontsize = 15)
plt.grid(True)

plt.xticks(rotation=45)

plt.savefig("Number of Earthquakes Since 1965.png")
plt.show()
```

```python
#create new dataframe to show highest earthquake magnitude for each year
maxmagnitude_year =df.groupby('year')['Magnitude'].max().reset_index()
maxmagnitude_year.head()
```

```python
#Graph the highest magnitude earthquake for each year
fig = plt.figure()
plt.figure(figsize=(20, 8))

plt.scatter(maxmagnitude_year['year'], maxmagnitude_year['Magnitude'])
plt.xlabel("1965-2016")
plt.ylabel("Magnitude of Earthquake")
plt.title("Highest Maqnitude of Earthquake per year Since 1965")

plt.xticks(rotation=45)
plt.grid(True)

plt.savefig('Highest Magnitued Earthquake per year Since 1965.png')
plt.show()
```

```python
#create new dataframe to show average earthquake magnitude for each year
meanmagnitude_year =df.groupby('year')['Magnitude'].mean().reset_index()
meanmagnitude_year.head()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))

plt.scatter(meanmagnitude_year['year'], meanmagnitude_year['Magnitude'])
plt.xlabel("1965-2016")
plt.ylabel("Average Magnitude of Earthquake")
plt.title("Average Maqnitude of Earthquake Since 1965")

plt.xticks(rotation=45)
plt.grid(True)

plt.savefig('Average magnitude of Earthquake Since 1965.png')
plt.show()
```
