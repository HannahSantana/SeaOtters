```python
import pandas as pd 
import matplotlib.pyplot as plt


df = pd.read_csv('1955-2017_hail.csv')
df.head()
```

```python
yearly_changes = df.groupby('yr').count().reset_index()
yearly_changes
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))


plt.scatter(yearly_changes.yr, yearly_changes.time)
```
