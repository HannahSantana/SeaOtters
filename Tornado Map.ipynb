```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create earthquake dataframe
tornado_df = pd.read_csv('1950-2017_actual_tornadoes_prelim.csv')
tornado_df.dropna()
tornado_df.head()
```

```python
# Store latitude and longitude in locations
locations = tornado_df[["slat", "slon"]]

# Fill NaN values and convert to float
rating = tornado_df["mag"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=6,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
