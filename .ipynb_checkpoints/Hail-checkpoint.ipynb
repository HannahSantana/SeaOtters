```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('1955-2017_hail.csv')
df.head()
```

```python
yearly_changes = df.groupby('yr').count().reset_index()
yearly_changes.head()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))
sns.regplot(yearly_changes['yr'], yearly_changes['time'])

plt.scatter(yearly_changes.yr, yearly_changes.time)
plt.xlabel('1955-2018', fontsize = 15)
plt.ylabel('Number of Hail Storms', fontsize = 15)
plt.title('Hail Storms since 1955', fontsize = 15)

plt.savefig('hail_storms_1955.png')
```

```python
mag = df.groupby('yr')['mag'].mean().reset_index()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))
sns.regplot(mag['yr'], mag['mag'])

plt.scatter(mag.yr, mag.mag)
plt.xlabel('1955-2018', fontsize = 15)
plt.ylabel('Average Diameter of Hail in inches', fontsize = 15)
plt.title('Average Diameter of Hail', fontsize = 15)

plt.savefig('avg_diameter_hail.png')
```

```python
maximum = df.groupby('yr')['mag'].max().reset_index()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))
sns.regplot(maximum['yr'], maximum['mag'])

plt.scatter(maximum.yr, maximum.mag)
plt.xlabel('1955-2018', fontsize = 15)
plt.ylabel('Maximum Diameter of Hail in inches', fontsize = 15)
plt.title('Maximum Diameter of Hail', fontsize = 15)

plt.savefig('max_diameter_of_hail.png')
```
