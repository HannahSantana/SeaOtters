```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('database.csv')
```

```python
df.head()
```

```python
df['year'] = df['Date'].apply(lambda x: x.split("/")[-1])
```

```python
df.head()
```

```python
# Create df to count earthquakes by year
df['count'] = 1
df1 = df[['year','count']]
earthquakes_by_year = df1.groupby(['year']).sum()['count'].reset_index()
earthquakes_by_year.head()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))

plt.scatter(earthquakes_by_year['year'], earthquakes_by_year['count'])
plt.xlabel("1965-2026")
plt.ylabel("Number of Earthquakes")
plt.title("Number of Earthquakes Since 1965")

plt.xticks(rotation=45)

plt.show()
```

```python
#create new dataframe to show highest earthquake magnitude for each year
maxmagnitude_year =df.groupby('year')['Magnitude'].max().reset_index()
maxmagnitude_year.head()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))

plt.scatter(maxmagnitude_year['year'], maxmagnitude_year['Magnitude'])
plt.xlabel("1965-2016")
plt.ylabel("Magnitude of Earthquake")
plt.title("Highest Maqnitude of Earthquake Since 1965")

plt.xticks(rotation=45)

plt.show()
```

```python
#create new dataframe to show average earthquake magnitude for each year
meanmagnitude_year =df.groupby('year')['Magnitude'].mean().reset_index()
meanmagnitude_year.head()
```

```python
fig = plt.figure()
plt.figure(figsize=(20, 8))

plt.scatter(meanmagnitude_year['year'], meanmagnitude_year['Magnitude'])
plt.xlabel("1965-2016")
plt.ylabel("Average Magnitude of Earthquake")
plt.title("Average Maqnitude of Earthquake Since 1965")

plt.xticks(rotation=45)

plt.show()
```
