```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Read in csv file
df = pd.read_csv('1950-2017_actual_tornadoes_prelim.csv')
```

```python
df.head()
```

```python
#Create df to count number of tornados by year
df['count'] = 1
df1 = df[['yr','count']]
tornados_by_year = df1.groupby(['yr']).sum()['count'].reset_index()
tornados_by_year.head()

#Create df for linear regression analysis
tornado_year = tornados_by_year['yr'].values[:,np.newaxis]
tornado_count = tornados_by_year['count'].values

#Intensity Analysis
tornadof1= df[(df['mag']==1)]
tornadof2= df[(df['mag']==2)]
```

```python
tornados_by_year.dtypes
```

```python
#Plot both frequency and regression
fig = plt.figure()
plt.figure(figsize=(20, 8))

lm = linear_model.LinearRegression()

lm.fit(tornado_year, tornado_count)

plt.scatter(tornado_year,tornado_count, color='b')
plt.plot(tornado_year, lm.predict(tornado_year),color='r')

plt.title("Tornado Frequency in the United States 1950 - 2017", fontsize = 15)
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Number of Tornado's", fontsize = 15)
plt.grid(True)

plt.savefig('Tornado Frequency in the United States 1950 - 2017')
plt.show()
```
